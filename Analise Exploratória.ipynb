```python
#importa os Datasets de treino e validação
data = pd.read_csv('customers.csv')
data = data.drop(['Channel' ,'Region'], axis = 1)
```

```python
data.head()
```

```python
fig, ax = plt.subplots( figsize= (20,10) )

for i, var in enumerate(data.columns):
    sns.kdeplot(np.log(data[var]), shade=True).set_xlim([-1,20])

```

```python
data = data[abs(data - np.mean(data)) < 2 * np.std(data)]
```

```python
plt.subplots( figsize= (20,10) )
for i, var in enumerate(data.columns):
    sns.violinplot(data = data, showfliers=False)
```

```python
plt.figure(figsize=(10, 10))
plt.title('Pearson correlation of continuous features', y=1.05, size=15)
sns.heatmap(data.corr(),linewidths=0.1,vmax=1.0, vmin= -1, square=True, linecolor='white', annot=True, cmap = 'RdBu_r' )
plt.show()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(15,10))
sns.regplot(x=data['Grocery'], y=data['Detergents_Paper'])
sns.regplot(x=data['Grocery'], y=data['Milk'], ax = ax[0])
```
